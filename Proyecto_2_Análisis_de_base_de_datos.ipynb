```python
paises_africa = [
    "Algeria", "Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", "Cabo Verde",
    "Cameroon", "Central African Republic", "Chad", "Comoros", "Democratic Republic of the Congo",
    "Republic of the Congo", "Ivory Coast", "Djibouti", "Egypt", "Equatorial Guinea",
    "Eritrea", "Eswatini", "Ethiopia", "Gabon", "Gambia", "Ghana", "Guinea", "Guinea-Bissau", "Kenya",
    "Lesotho", "Liberia", "Libya", "Madagascar", "Malawi", "Mali", "Mauritania", "Mauritius", "Morocco",
    "Mozambique", "Namibia", "Niger", "Nigeria", "Rwanda", "Sahara Occidental", "Senegal", "Seychelles",
    "Sierra Leone", "Somalia", "South Africa", "Sudan", "South Sudan", "Tanzania", "Togo", "Tunisia",
    "Uganda", "Zambia", "Zimbabwe"
]
df_africa = df[df['COUNTRY'].isin(paises_africa)]
print(df_africa)
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
fecha_especifica = '2020-03-15' #Fecha arbitraria dentro del rango de fechas
df_fecha = df[df['DATE'] == fecha_especifica]
df_africa_fecha = df_fecha[df_fecha['COUNTRY'].isin(paises_africa)] #Filtramos el DataFrame para obtener solo los datos de esa fecha

media_gdpcap = df_africa_fecha['GDPCAP'].mean() #Calculamos la media del PIB per cápita para los países africanos disponibles en esa fecha
pais_alto = df_africa_fecha.loc[df_africa_fecha['GDPCAP'].idxmax()] #Seleccionamos el país con el PIB per cápita más alto
pais_bajo = df_africa_fecha.loc[df_africa_fecha['GDPCAP'].idxmin()] #Seleccionamos el país con el PIB per cápita más bajo
df_africa_fecha['Diferencia_media'] = (df_africa_fecha['GDPCAP'] - media_gdpcap).abs() #Seleccionamos el país cuyo PIB per cápita esté más cercano a la media
pais_cercano_media = df_africa_fecha.loc[df_africa_fecha['Diferencia_media'].idxmin()]

print(pais_alto[['COUNTRY', 'GDPCAP']])
print(pais_bajo[['COUNTRY', 'GDPCAP']])
print(pais_cercano_media[['COUNTRY', 'GDPCAP']])
```

```python
code = df["COUNTRY"].unique().tolist() #obtiene los valores únicos de la columna "COUNTRY", estos valores únicos se almacenan en la lista code.
hdi = [] #Crear listas vacías que se utilizarán más adelante para almacenar los valores agregados (sumados) de ciertas columnas del DataFrame
tc = []
td = []
sti = []
population = df["POP"].unique().tolist()
gdp = []

mod_value = 210
for value in code: #Este bucle recorre cada valor único de code, que corresponde a cada grupo de datos en el DataFrame. Para cada valor de value.
    hdi.append((df.loc[df["COUNTRY"] == value, "HDI"]).sum()/mod_value)
    tc.append((df.loc[df["COUNTRY"] == value, "TC"]).sum())
    td.append((df.loc[df["COUNTRY"] == value, "TD"]).sum())
    sti.append((df.loc[df["COUNTRY"] == value, "STI"]).sum()/mod_value)
    population.append((df.loc[df["COUNTRY"] == value, "POP"]).sum()/mod_value)

aggregated_df = pd.DataFrame(
    list(zip(code, hdi, tc, td, sti, population)),
    columns = ["COUNTRY", "HDI",
               "TC", "TD", "STI", "POP"]
)

aggregated_df.head()
```
